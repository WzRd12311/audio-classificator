```python
import os
from pathlib import Path
from tqdm.notebook import tqdm
from concurrent.futures import ProcessPoolExecutor
from my_utils import process_file
```

```python
BASE_DIR = Path(r'C:\audio_classificator\dataset')
CONFIG = [
    {
        'input': BASE_DIR / 'myaudio' / 'human',
        'output': BASE_DIR / 'chunks' / 'human'
    },
    {
        'input': BASE_DIR / 'myaudio' / 'bot',
        'output': BASE_DIR / 'chunks' / 'bot'
    }
]
```

Функция для обработки одного файла. Запускается в отдельном процессе.

```python
def main():
    tasks = []
    
    for conf in CONFIG:
        input_dir = conf['input']
        output_dir = conf['output']
        
        if input_dir.exists():
            files = [f for f in input_dir.iterdir() if f.is_file()]
            for f in files:
                tasks.append((f, output_dir))
        else:
            print(f"Warning: Directory not found: {input_dir}")

    print(f"Found {len(tasks)} files to process. Starting pool...")

    with ProcessPoolExecutor() as executor:
        list(tqdm(executor.map(process_file, tasks), total=len(tasks)))

    print("Done!")
```

```python
if __name__ == '__main__':
    main()
```
